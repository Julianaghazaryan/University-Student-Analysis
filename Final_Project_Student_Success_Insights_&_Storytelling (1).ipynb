# Final Project: Student Success Insights & Storytelling
### Project Overview
The goal of this assignment is to transition from raw data exploration to a focused narrative. By analyzing academic habits such as study hours, sleep patterns, and attendance, we aim to identify the key drivers of student success (GPA).


```python
import pandas as pd
import numpy as np
import plotly.express as px

# Loading the dataset
df = pd.read_csv('Student_data.csv')

# Data Cleaning: Removing non-statistical columns and handling missing values
if 'Student_ID' in df.columns:
    df = df.drop(columns=['Student_ID'])

df = df.dropna()

print(f"Data refined. Current shape: {df.shape}")
```

### Insight 1: How Study and Sleep Affect GPA
In this visualization, we explore the "Golden Ratio" between studying and resting. We want to see if more study hours always lead to a higher GPA, or if sleep plays a critical role.

```python
# 3D Scatter plot to show the relationship between Study, Sleep, and GPA
fig1 = px.scatter(df,
                 x='Study_Hours_Per_Day',
                 y='Sleep_Hours',
                 color='Final_CGPA',
                 size='Attendance_Pct',
                 hover_data=['Major'],
                 title='The Impact of Study and Sleep Balance on GPA',
                 labels={'Study_Hours_Per_Day': 'Daily Study Hours', 'Sleep_Hours': 'Daily Sleep Hours'},
                 template='plotly_dark')
fig1.show()
```

### Insight 2: Average GPA by Major
Does the field of study influence the average grade? This chart highlights which majors have the highest academic performance, providing a baseline for our dashboard's filtering system.

```python
# Calculating average GPA per Major
major_stats = df.groupby('Major')['Final_CGPA'].mean().sort_values(ascending=False).reset_index()

fig2 = px.bar(major_stats,
              x='Major',
              y='Final_CGPA',
              color='Final_CGPA',
              title='Average Academic Performance by Major',
              text_auto='.2f',
              color_continuous_scale='Blues')
fig2.show()
```

### Insight 3: Correlation Matrix
To build an effective predictive model later, we need to know which features are most closely linked to the GPA. The correlation matrix helps us "select what actually matters."

```python
# Heatmap for correlation
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

fig3 = px.imshow(corr_matrix,
                 text_auto='.2f',
                 title='Correlation Matrix of Academic Success Factors',
                 color_continuous_scale='RdBu_r')
fig3.show()
```

### Insight 2: Academic Performance Distribution by Major
While average GPA gives us a general idea, it's important to look at the **distribution** and **spread** of grades within each major. This Box Plot helps us identify:
* **Outliers:** Are there students performing significantly better or worse than their peers?
* **Consistency:** Which majors have the most consistent grading patterns?
* **Comparison:** How does the median GPA of a STEM major compare to others?

```python
fig4 = px.box(df, x='Major', y='Final_CGPA', points="all",
             title='GPA Distribution by Major',
             color='Major')
fig4.show()
```

*Analysis:* The box plot shows that while most majors have a similar median GPA, the range of grades varies. This insight is crucial for our dashboard to ensure that users are compared against the specific distribution of their own department rather than a generic average.

### Final Story & Dashboard Strategy
Based on the visual evidence, our data tells a clear story:
1. **Consistency over Intensity:** High attendance and balanced sleep appear more closely linked to top-tier GPAs than extreme study hours alone.
2. **Predictable Patterns:** Certain academic behaviors strongly correlate with success across all majors.

**Next Steps for the Dashboard:**
The final dashboard will be designed as a **Success Predictor**. Users will be able to input their daily habits (study/sleep) and see where they stand compared to the average in their specific Major.
